```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import sklearn.cluster as cluster
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/IRIS.csv')
df.head()
```

```python
df=df.drop('species',axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python

scaler = MinMaxScaler()
scale = scaler.fit_transform(df)
df_scale = pd.DataFrame(scale, columns = [df]);
df_scale.head(5)
```

```python
km=KMeans(n_clusters=3)
y_predicted = km.fit_predict(df_scale)
y_predicted
```

```python
km.cluster_centers_
```

```python
df['Clusters'] = km.labels_
sns.scatterplot(x="petal_length", y="petal_width",hue = 'Clusters',  data=df,palette='viridis')
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width",hue = 'Clusters',  data=df,palette='viridis')
```

```python
sns.scatterplot(x="petal_width", y="sepal_length",hue = 'Clusters',  data=df,palette='viridis')
```

```python

K=range(2,12)
wss = []

for k in K:
    kmeans=cluster.KMeans(n_clusters=k)
    kmeans=kmeans.fit(df_scale)
    wss_iter = kmeans.inertia_
    wss.append(wss_iter)
```

```python

plt.xlabel('K')
plt.ylabel('Within-Cluster-Sum of Squared Errors (WSS)')
plt.plot(K,wss)
     
```
